```python
import pandas as pd
import numpy as np

# GitHub의 자동차 생산 자료를 읽어들인다
path = 'https://github.com/dongupak/DataML/raw/main/csv/'
file = path + 'vehicle_prod.csv'
df = pd.read_csv(file, index_col=0)

# 각 행의 평균을 계산하여 'Average' 열 추가
df['Average'] = df.select_dtypes(np.number).mean(axis=1)

# 데이터 프레임에서 국가명을 제외한 숫자의 합을 구하여 Total로 명명
sums = df.select_dtypes(np.number).sum().rename('Total')

# pd.concat()을 사용하여 Total 행을 데이터프레임에 추가
total_df = pd.concat([df, pd.DataFrame([sums])])
total_df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = 'https://github.com/dongupak/DataML/raw/main/csv/'
weather_file = path + 'weather.csv'
weather = pd.read_csv(weather_file, encoding='CP949')

# '일시' 열에서 month를 추출하여 'month' 열 추가 (오타 수정)
weather['month'] = pd.DatetimeIndex(weather['일시']).month

# 월별 데이터와 월별 평균 기온을 저장할 리스트 초기화 (오타 수정)
monthly = [None for x in range(12)]
monthly_wind = [0 for x in range(12)]

# 월별 평균 기온 계산
for i in range(12):
    monthly[i] = weather[weather['month'] == i + 1]
    monthly_wind[i] = monthly[i]['평균기온'].mean()

# 그래프 그리기
months = np.arange(1, 13)
plt.bar(months, monthly_wind, color='green')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.title('Average Monthly Temperature')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# GitHub에서 날씨 데이터를 읽어들인다
path = 'https://github.com/dongupak/DataML/raw/main/csv/'
weather_file = path + 'weather.csv'
weather = pd.read_csv(weather_file, encoding='CP949')

# '일시' 열에서 month를 추출하여 'month' 열 추가
weather['month'] = pd.DatetimeIndex(weather['일시']).month

# 월별 데이터와 월별 평균 풍속을 저장할 리스트 초기화
monthly = [None for x in range(12)]
monthly_wind = [0 for x in range(12)]

# 월별 평균 풍속 계산
for i in range(12):
    monthly[i] = weather[weather['month'] == i + 1]
    monthly_wind[i] = monthly[i]['평균풍속'].mean()

# 그래프 그리기
months = np.arange(1, 13)
plt.bar(months, monthly_wind, color='blue')
plt.xlabel('Month')
plt.ylabel('Average Wind Speed (m/s)')
plt.title('Average Monthly Wind Speed')
plt.show()
```
