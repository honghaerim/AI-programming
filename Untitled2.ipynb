```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 랜덤 데이터 생성
m = 150  # 데이터 포인트 수
x = 12 * np.random.rand(m, 1) - 6
y = 0.5 * x**2 + 2 * x + 1 + np.random.randn(m, 1)


# 다항 특성 생성
poly_features = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly_features.fit_transform(x)

# 모델 적합
regr = LinearRegression()
regr.fit(x_poly, y)

# 다항 회귀선을 위한 값 생성
domain = np.linspace(-6, 6, 150).reshape(-1, 1)
domain2 = poly_features.fit_transform(domain)
y_predict = regr.predict(domain2)

# 그래프 그리기
plt.figure(figsize=(8, 5))
plt.scatter(x, y)
plt.plot(domain, y_predict, color='red', linewidth=4)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Polynomial Regression")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# 닥스훈트의 길이와 높이 데이터 (10마리)
dach_length = [55, 57, 64, 63, 58, 49, 54, 61, 60, 59]
dach_height = [30, 31, 36, 30, 33, 25, 37, 34, 29, 32]

# 진돗개의 길이와 높이 데이터 (10마리)
jin_length = [56, 47, 56, 46, 49, 53, 52, 48, 55, 51]
jin_height = [52, 52, 50, 53, 50, 53, 49, 54, 51, 55]

# 특성 데이터 (길이, 높이)
X = np.array([[length, height] for length, height in zip(dach_length + jin_length, dach_height + jin_height)])
# 레이블 (0: Dachshund, 1: Jindo dog)
y = np.array([0]*len(dach_length) + [1]*len(jin_length))

# 새로운 데이터
newdata_length = [54]
newdata_height = [43]
new = np.array([[54, 43]])

# k-NN 분류기 모델 생성
k = 5
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X, y)

# 새로운 데이터의 클래스 예측
prediction = knn.predict(new)

# 산점도 그리기
plt.figure(figsize=(8, 5))
plt.scatter(dach_length, dach_height, c='red', marker='o', label='Dachshund')  # 닥스훈트: 빨간색 원
plt.scatter(jin_length, jin_height, c='blue', marker='^', label='Jindo dog')   # 진돗개: 파란색 삼각형
plt.scatter(newdata_length, newdata_height, c='green', marker='s', s=100, label='New Data')  # 새로운 데이터: 녹색 사각형

# 예측 결과 표시
plt.xlabel('Length')
plt.ylabel('Height')
plt.title('Dog Size Classification using k-NN')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
