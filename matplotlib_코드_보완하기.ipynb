```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3, 3)

grid = plt.GridSpec(3, 3, wspace=0.4, hspace=0.3)
X = np.random.randn(100) # 정규 분포를 가지는 데이터
Y = np.random.randn(100) # 정규 분포를 가지는 데이터
plt.subplot(grid[0, :]).scatter(X, Y)  #산점도 그림

X = np.linspace(0, 10, 100)
Y = np.cos(X)
plt.subplot(grid[2, 0]).plot(X, Y, color='g')

X = np.arange(10)
Y = np.abs(np.random.randn(10) * 10)
plt.subplot(grid[2, 1]).bar(X, Y)

Z = np.random.uniform(0, 1, (10, 5))
plt.subplot(grid[1:, 2]).imshow(Z)
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3, 3)

grid = plt.GridSpec(3, 3, wspace=0.4, hspace=0.3)

X = np.random.randn(100) # 정규 분포를 가지는 데이터
Y = np.random.randn(100) # 정규 분포를 가지는 데이터
plt.subplot(grid[0, :]).scatter(X, Y)  #산점도 그림

X = np.linspace(0, 10, 100)
Y = np.cos(X)
plt.subplot(grid[2, 0]).plot(X, Y, color='g')

X = np.arange(10)
Y = np.abs(np.random.randn(10) * 10)
plt.subplot(grid[2, 1]).bar(X, Y)

Z = np.random.uniform(0, 1, (10, 5))
plt.subplot(grid[1:, 2]).imshow(Z)

X = np.linspace(0, 4, 100)
Y = X
plt.subplot(grid[1, 0:2]).plot(X, Y, color='r')

plt.show()
```
