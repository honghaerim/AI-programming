```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 랜덤 데이터 생성
m = 50  # 데이터 포인트 수
x = 8 * np.random.rand(m, 1) - 4
y = 0.5 * x**2 + 0.5 * x + 1 + np.random.randn(m, 1)

# 다항 특성 생성
poly_features = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly_features.fit_transform(x)

# 모델 적합
regr = LinearRegression()
regr.fit(x_poly, y)

# 다항 회귀선을 위한 값 생성
domain = np.linspace(-4, 4, 50).reshape(-1, 1)
domain2 = poly_features.fit_transform(domain)
y_predict = regr.predict(domain2)

# 그래프 그리기
plt.figure(figsize=(5, 3))
plt.scatter(x, y)
plt.plot(domain, y_predict, color='yellow', linewidth=3)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Polynomial Regression")
plt.show()
```
